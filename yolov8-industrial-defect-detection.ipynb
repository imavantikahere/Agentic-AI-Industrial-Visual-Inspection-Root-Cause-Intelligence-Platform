```python
#EDA
import os
from collections import Counter
import matplotlib.pyplot as plt

labels_path = f"/kaggle/working/Industrial-defect-detection-1/train/labels"

class_counts = Counter()

for file in os.listdir(labels_path):
    if file.endswith(".txt"):
        with open(os.path.join(labels_path, file), "r") as f:
            for line in f:
                class_id = int(line.split()[0])
                class_counts[class_id] += 1

print(class_counts)

plt.bar(class_counts.keys(), class_counts.values())
plt.xlabel("Class ID")
plt.ylabel("Count")
plt.title("Defect Class Distribution")
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

labels_path = f"/kaggle/working/Industrial-defect-detection-1/train/labels"

widths = []
heights = []

for file in os.listdir(labels_path):
    if file.endswith(".txt"):
        with open(os.path.join(labels_path, file)) as f:
            for line in f:
                parts = list(map(float, line.strip().split()))

                coords = parts[1:]  # skip class id

                x_vals = coords[0::2]
                y_vals = coords[1::2]

                bbox_width = max(x_vals) - min(x_vals)
                bbox_height = max(y_vals) - min(y_vals)

                widths.append(bbox_width)
                heights.append(bbox_height)

plt.figure(figsize=(6,6))
plt.scatter(widths, heights, alpha=0.4)
plt.xlabel("Approx Bounding Box Width")
plt.ylabel("Approx Bounding Box Height")
plt.title("Defect Region Size Distribution")
plt.show()
```
